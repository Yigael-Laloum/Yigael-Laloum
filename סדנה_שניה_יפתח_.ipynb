```python
#ככה נראה מבנה של קוד הכי בסיסי בפייתון

#התקנת חבילות
!pip install spacy

#ייבוא ספריות
import spacy

# Load the English model from spaCy.
nlp = spacy.load("en_core_web_sm")

#פונקציה שסופרת מילים ושמות
def process_text(text):
    # Process the text using spaCy.
    doc = nlp(text)

    # Initialize a counter for the words and a list for the names.
    word_count = 0
    names = []

    # Loop through the words (tokens) in the processed text.
    for token in doc:
        # Count each token if it is a word (not punctuation).
        if not token.is_punct:
            word_count += 1

    # Loop through the recognized entities in the text.
    for ent in doc.ents:
        # If the entity is a proper name (PERSON), add it to the names list.
        if ent.label_ == "PERSON":
            names.append(ent.text)

    # Return the word count and the list of names.
    return word_count, names

# פונקציה מרכזית הגדרות והפעלה
def main():
    # טקסט דוגמא
    text = """
    John and Mary went to the park to meet with Michael and Sarah.
    They were talking about a new project that Elizabeth and James were working on.
    """

    # הפעלת הפונקציה של הספירה
    word_count, names = process_text(text)

    # הדפסת התוצאות
    print(f"Total word count (excluding punctuation): {word_count}")
    print(f"Names found in the text: {', '.join(names)}")

# הגדרת והפעלת הפונקציה הראשית
if __name__ == "__main__":
    main()
```
